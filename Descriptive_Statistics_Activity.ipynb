# Descriptive Statistics Activity

Welcome to this interactive activity! In this notebook, you will explore **basic descriptive statistics** using a dataset of student test scores. Let's get started!

```python
import pandas as pd

file_path = 'Student_Test_Scores.csv'  
df = pd.read_csv(file_path)

df
```

## 1. Calculate Basic Descriptive Statistics

```python
# Mean
mean_score = df['Test Score'].mean()
print(f'Mean (Average) Score: {mean_score}')

# Median
median_score = df['Test Score'].median()
print(f'Median Score: {median_score}')

# Mode
mode_score = df['Test Score'].mode()[0]
print(f'Mode Score: {mode_score}')
```

## 2. Calculate Range and Standard Deviation

```python
# Range
range_score = df['Test Score'].max() - df['Test Score'].min()
print(f'Range: {range_score}')

# Standard Deviation
std_dev = df['Test Score'].std()
print(f'Standard Deviation: {std_dev}')
```

## 3. Visualize the Data 

```python
import matplotlib.pyplot as plt

plt.bar(df['Name'], df['Test Score'], color='skyblue')
plt.xlabel('Student Name')
plt.ylabel('Test Score')
plt.title('Student Test Scores')
plt.xticks(rotation=45)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Calculate the mean and standard deviation of the test scores
mean_score = df['Test Score'].mean()
std_dev = df['Test Score'].std()

# Generate the x-axis values (test score range)
x = np.linspace(df['Test Score'].min(), df['Test Score'].max(), 100)

# Generate the y-axis values for the normal distribution
y = norm.pdf(x, mean_score, std_dev)

# Plot the histogram of the test scores
plt.hist(df['Test Score'], bins=5, density=True, alpha=0.6, color='skyblue', edgecolor='black')

# Plot the normal distribution curve
plt.plot(x, y, color='red', linewidth=2, label='Normal Distribution')

# Add labels and title
plt.title('Test Scores with Normal Distribution Bell Curve')
plt.xlabel('Test Score')
plt.ylabel('Density')
plt.legend()

# Show the plot
plt.show()
```
